**1) Let's import some usefull modules**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**2) Create random data using numpy closures**

```python
data = np.random.normal(loc=0, scale=1, size=(100,2))
print(f"First 5 rows of data:\n\n {data[:5,:]}")
```

**3) Basic visualization of data with the use of matplotlib module**

```python
style_list = ['default', 'classic'] + sorted(
        style for style in plt.style.available
        if style != 'classic' and not style.startswith('_'))

for st in style_list:
  with plt.style.context(st):
    fig, ax = plt.subplots(figsize=(8,4))
    # fig.tight_layout()
    ax.plot(data[:,0], data[:,1], color='b', linestyle='none', marker='o', markersize=10, alpha=0.3)
    ax.scatter(data[:,0], data[:,1], color='r', alpha=0.3)
    fig.suptitle(st, x=0.01, ha='left', fontsize=14, fontfamily='DejaVu Sans', 
                 fontweight='normal')
```
